## Data Wrangling

```python
# Renaming Columns
df.rename(columns = {"Happiness score": "Happiness_Score", "Explained by: GDP per capita": "GDP", "Explained by: Social support": "Social_Support", 
          "Explained by: Healthy life expectancy": "Life_Expectancy", 
          "Explained by: Freedom to make life choices": "Life_Choices", "Explained by: Generosity": "Generosity", 
          "Explained by: Perceptions of corruption": "Perceptions_of_Corruption"}, inplace=True)
df
```

```python
# Using regex to get rid of the asterisk next to a few of the countries' names

df.replace("\*", '', regex=True, inplace=True)
df
```

```python
results = df.groupby(["Life_Expectancy", "Country"]).agg({"Happiness_Score": ["mean", "min", "max"]})
print("Mean, min, and max values of Happiness Score grouped by Country")
print(results)
```

```python
df["Life_Expectancy"].mean()
```

```python
# filtering for Life Expectancy of Countries above mean

df2 = df.loc[df["Life_Expectancy"] > (df["Life_Expectancy"].mean())]
df2
```

```python
# filtering for Life Expectancy of Countries under mean
df3 = df.loc[df["Life_Expectancy"] < (df["Life_Expectancy"].mean())]
df3
```

### Matplotlib Line Plot

```python
plt.figure(figsize=(10, 10)) # configuring figure size of plot
plt.scatter(df2.Happiness_Score, df2.Life_Expectancy, label="Over Age 59 (mean)", color="red") 
plt.scatter(df3.Happiness_Score, df3.Life_Expectancy, label="Under Age 59 (mean)", color="blue")
plt.title("Comparing Life Expectancy and Happiness Score by Average Age", fontsize=16) # plot title, changing size of title
plt.xlabel("Happiness Score", fontsize= 16) # x-label, changing size of axis label
plt.ylabel("Life Expectancy", fontsize= 16) # y-label, changing size of axis label
plt.legend(bbox_to_anchor=(1.50,1), fontsize=15) # Changing Legend position and placing outside of plot
plt.show()
```

When plotting matplotlib, plt.show() must be used in order to display the plots. When using Seaborn, plt.show() isn't necessary to use to display the plots.
